```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup("https://www.sharesansar.com/live-trading", 'html.parser')
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text, "lxml")
code
```

```python
table_code = code.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
list_data = []
for i in tags:
    data = i.text
    list_data.append(data)

print(list_data)
```

```python
cleaned_data = []
for i in list_data[1:]:
    x = (i.replace("                                        ",""))
    x = x.split('\n\n')
    x = [j .replace('\n','') for j in x]
    cleaned_data.append(x[1:])

cleaned_data
```

```python
col = list_data[0].split('\n')[1:-1]
col
```

```python
import pandas as pd
df = pd.DataFrame(cleaned_data,columns = col)
df
```

```python
df['LTP'] = [float(i.replace(',','')) for i in df['LTP']]
df['Point Change'] = [float(i) for i in df['Point Change']]
df['% Change'] = [float(i) for i in df['% Change']]
df['Open'] = [float(i.replace(',','')) for i in df['Open']]
df['High'] = [float(i.replace(',','')) for i in df['High']]
df['Low'] = [float(i.replace(',','')) for i in df['Low']]
df['Volume'] = [float(i.replace(',','')) for i in df['Volume']]
df['Prev. Close'] = [float(i.replace(',','')) for i in df['Prev. Close']]
```

```python
df
```

```python
import plotly.express as px
fig = px.bar(df, x='Symbol', y='Point Change')
fig.show()
```

```python
# top_10_positive_change = df[df['% Change'] > 0].nlargest(10, '% Change')
top_10_positive_change = df.sort_values(by='% Change', ascending=False).head(10)
fig = px.bar(top_10_positive_change, x='Symbol', y='% Change')
fig.show()
top_10_positive_change
```

```python
# top_10_positive_change = df[df['% Change'] > 0].nlargest(10, '% Change')
top_10_neg_change = df.sort_values(by='% Change', ascending=True).head(10)
fig = px.bar(top_10_neg_change, x='Symbol', y='% Change')
fig.show()
```

```python
# top_10_positive_change = df[df['% Change'] > 0].nlargest(10, '% Change')
top_10_neg_change = df.sort_values(by='Volume', ascending=False).head(10)
fig = px.bar(top_10_neg_change, x='Symbol', y='Volume')
fig.show()
```

```python
top_change = df.sort_values(by='% Change', ascending=False).head(10)
top_volume = top_change.sort_values(by='Volume', ascending=False).head(10)
fig = px.bar(top_volume, x='Symbol', y='Volume')
fig.show()
```

```python
import plotly.express as px
top_10_volume = df.sort_values(by='Volume', ascending=False).head(10)
fig = px.pie(top_10_volume, values='Volume', names='Symbol', title='Top 10 Volume Analysis')
fig.show()
```
